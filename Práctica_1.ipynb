```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_bloch_multivector
```

## - psi0
Compuertas: 0

```python
psi = QuantumCircuit(1)
psi.draw('mpl')
```

```python
psi_0 = Statevector(psi)
print("➤ en forma de vector:")
display(array_to_latex(psi_0.data))
print("➤ en forma de ket:")
display(psi_0.draw('latex'))
print("➤ esfera de Bloch")
display(plot_bloch_multivector(psi))
```
