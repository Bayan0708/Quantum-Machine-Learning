```python
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
```

# **Algoritmo de Deutsch**

image.png

```python
s_circuit = QuantumCircuit(2,1)
s_circuit.x(0)
s_circuit.barrier()
s_circuit.h(0)
s_circuit.h(1)
s_circuit.barrier()

s_circuit.draw('mpl')
```

## 1) Constante: f(x) = 0.


```python
# Constante: f(x) = 0
oracle_const_2 = QuantumCircuit(2, name="f(x)=0")
oracle_const_2.barrier()

oracle_const_2.draw('mpl')
```

```python
Uf_const_2 = s_circuit.compose(oracle_const_2)
Uf_const_2.h(1)
Statevector(Uf_const_2).draw('latex', max_size = 16, prefix="\\vert\\psi_3\\rangle = ")
```

```python
Uf_const_2.measure(1, 0)
Uf_const_2.draw('mpl')
```

```python
shots = 100
simulator = AerSimulator()
compiled_circuit_1 = transpile(Uf_const_2, simulator)
job = simulator.run(compiled_circuit_1, shots=shots)
result = job.result()
counts = result.get_counts()
plot_histogram(counts, title="Resultados de Deutsch")
```

## 2) Balanceado: f(x) = x.


```python
# Balanceado: f(x) = x
oracle_bal_1 = QuantumCircuit(2, name="f(x)=x")
oracle_bal_1.cx(0, 1)
oracle_bal_1.barrier()

oracle_bal_1.draw('mpl')
```

```python
Uf_bal_1 = s_circuit.compose(oracle_bal_1)
Uf_bal_1.h(1)
Statevector(Uf_bal_1).draw('latex', max_size = 16, prefix="\\vert\\psi_3\\rangle = ")
```

```python
Uf_bal_1.measure(1, 0)
Uf_bal_1.draw('mpl')
```

```python
shots = 100
simulator = AerSimulator()
compiled_circuit_2 = transpile(Uf_bal_1, simulator)
job = simulator.run(compiled_circuit_2, shots=shots)
result = job.result()
counts = result.get_counts()
plot_histogram(counts, title="Resultados de Deutsch")
```
